```python
from typing import ItemsView
from pandas.core.frame import DataFrame
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
import requests
```

```python
keywords=['anti islamic funny memes','memes about Prophet muhammad','funny photots against Islam','funny photos against quran',
'memes against muslim women','charlie hebdo cartoon about prophet','attacking cartoons about islam','islamophobia cartoons']
```

```python

def search_google(search_query):
    converted_name = search_query.lower().replace(" ", "_")
    data_dic = {
        'keywords': [],
        'Imagelink': []
    }

    search_url = f"https://www.google.com/search?site=&tbm=isch&source=hp&biw=1873&bih=990&q={search_query}"
    browser = webdriver.Chrome()
    browser.get(search_url)
    browser.maximize_window()
    time.sleep(2)

    scroll_pause_time = 2  # Time to wait for loading new images
    count = 0
    last_height = browser.execute_script("return document.body.scrollHeight")

    while True:
        # Find all image elements
        img_elements = browser.find_elements(By.CSS_SELECTOR, 'img.YQ4gaf')

        for img in img_elements[count:]:  # Start from the last position
            try:
                link = img.get_attribute('src')
                if link and link.startswith('http'):
                    data_dic['keywords'].append(search_query)
                    data_dic['Imagelink'].append(link)
                    count += 1

                    # Save data every 20 images
                    if count % 20 == 0:
                        df = pd.DataFrame(data=data_dic)
                        df.to_excel(f'excel_links/{converted_name}_links.xlsx', index=False)
                        print(f"Downloaded {count} images of {search_query}")

            except Exception as e:
                print(f'An error occurred: {e}')

        # Scroll down to load more images
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause_time)

        # Check if scrolling down loaded new images
        new_height = browser.execute_script("return document.body.scrollHeight")
        if new_height == last_height:  # If height didn't change, end the loop
            break
        last_height = new_height

    # Save final data
    df = pd.DataFrame(data=data_dic)
    df.to_excel(f'excel_links/{converted_name}_links.xlsx', index=False)
    print(f"Finished downloading images of {search_query}")

    browser.close()
```

```python
import urllib
query_list = [name.lower().replace(" ", "_") for name in keywords]
count=0

for query in query_list:
    url_data = pd.read_excel(f'excel_links/{query}_links.xlsx')
    url_list = list(url_data['Imagelink'])

    
    for url in url_list:
      
     try:
        with urllib.request.urlopen(url) as response:
            image_data=response.read()
            with open(f"images_all/{count}.jpg",'wb') as f:
                f.write(image_data)
                count+=1
        
     except urllib.error.HTTPError as e:
            print(f"Failed to download image from {url}. HTTP Error: {e.code}")
     except urllib.error.URLError as e:
            print(f"Failed to download image from {url}. URL Error: {e.reason}")
     except Exception as e:
            print(f"An unexpected error occurred while downloading {url}: {e}")
        
        
    print(f"downloaded image for query {query}")
    

    

    

    
```
